```python
def parse_sentence_term(path, lowercase=False):
    tree = parse(path)
    sentences = tree.getroot()
    data = []
    split_char = '__split__'
    for sentence in sentences:
        text = sentence.find('text')
        if text is None:
            continue
        text = text.text
        if lowercase:
            text = text.lower()
        aspectTerms = sentence.find('aspectTerms')
        if aspectTerms is None:
            continue
        for aspectTerm in aspectTerms:
            term = aspectTerm.get('term')
            if lowercase:
                term = term.lower()
            polarity = aspectTerm.get('polarity')
            start = aspectTerm.get('from')
            end = aspectTerm.get('to')
            piece = text + split_char + term + split_char + polarity + split_char + start + split_char + end
            data.append(piece)
    return data
```

```python
import stanza

nlp = stanza.Pipeline(lang='en', processors='tokenize,mwt,pos,lemma,depparse')

def parse_sentence(sentence, aspectTerms):
  doc = nlp(sentence)
  token = []
  pos = []
  head = []
  deprel = []
  aspects = []
  cnt = 0
  lookup = {}
  for sent in doc.sentences:
    for word in sent.words:
      token.append(word.text)
      pos.append(word.xpos)
      head.append(word.head)
      deprel.append(word.deprel)
      lookup[word.text] = cnt
      cnt += 1
  
  aspects = []
  if aspectTerms is None:
      aspects = []
  else:
    for aspectTerm in aspectTerms:
        term0 = aspectTerm.get('term').split()
        term = []
        for i in term0:
          if i not in lookup:
            if '-' in i:
              parts=i.split('-')
              for j in range(len(parts)-1):
                term.append(parts[j])
                term.append('-')
              term.append(parts[-1])
            if "'s" in i:
              parts=i.split("'s")
              for j in range(len(parts)-1):
                term.append(parts[j])
                term.append("'s")
            if '/' in i:
              parts=i.split('/')
              for j in range(len(parts)-1):
                term.append(parts[j])
                term.append('/')
              term.append(parts[-1])
          else:
            term.append(i)

        print(lookup)
        print(term0)
        print(term)
        print(term[0])
        frm = lookup[term[0]]
        to = lookup[term[-1]]+1
        if frm >= to:
          frm = to - len(term)
        polarity = aspectTerm.get('polarity')
        aspects.append({'term': term, 'from': frm, 'to': to, 'polarity': polarity})
  return {'token': token, 'pos': pos, 'head': head, 'deprel':deprel, 'aspects': aspects}
#parse_sentence('Set up was easy.')
```
